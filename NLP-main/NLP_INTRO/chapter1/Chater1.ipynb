```
# 01-4-1. 판다스(Pandas)
import pandas as pd
print(pd.__version__)

# 1) 시리즈(Series)
sr = pd.Series([17000, 18000, 1000, 5000],
               index=["피자", "치킨", "콜라", "맥주"])
print("sr ------------------------------")
print(sr)
print()
print("sr.values -----------------------")
print(sr.values)
print()
print("sr.index ------------------------")
print(sr.index)
print()
print()

# 2) 데이터프레임(DataFrame)
values = [[1,2,3], [4,5,6], [7,8,9]]
index = ['one', 'two', 'three']
columns = ['A', 'B', 'C']

df1 = pd.DataFrame(values, index=index, columns=columns)
print("df1 ------------------------------")
print(df1)
print()
print("df1.index -----------------------")
print(df1.index)
print()
print("df1.columns -----------------------")
print(df1.columns)
print()
print("df1.values -----------------------")
print(df1.values)
print()

# 3) 데이터프레임의 생성

## 리스트로 생성하기
print("[리스트로 생성]--------------------")

data2 = [
        ['1000', 'Steve', 90.72],
        ['1001', 'James', 78.09],
        ['1002', 'Doyeon', 98.43],
        ['1003', 'Jane', 64.19],
        ['1004', 'Pilwoong', 81.30],
        ['1005', 'Tony', 99.14],
]
df2 = pd.DataFrame(data2)
print("df2 ------------------------------")
print(df2)
print()

df2 = pd.DataFrame(data2, columns=['학번', '이름', '점수'])
print("df2 columns 지정------------------")
print(df2)
print()

print("[딕셔너리로 생성]--------------------")
data3 = {
         '학번': ['1000', '1001', '1002', '1003', '1004', '1005'],
         '이름': ['Steve', 'James', 'Doyeon', 'Jane', 'Pilwoong', 'Tony'],
         '점수': [90.72, 78.09, 98.43, 64.19, 81.30, 99.14]
}
df3 = pd.DataFrame(data3)
print("df3 ------------------------------")
print(df3)
print()

# 4) 데이터프레임 조회하기

# df.head(n)   - 앞 부분을 n개만 보기 
# df.tail(n)   - 뒷 부분을 n개만 보기 
# df['열이름'] - 해당되는 열을 확인

print("df.head(n) -----------------------")
print(df3.head(3))
print()

print("df.tail(n) -----------------------")
print(df3.tail(3))
print()

print("df['열이름'] ---------------------")
print(df3['학번'])
print()

# 5) 외부 데이터 읽기
import io
from google.colab import files
uploaded = files.upload()

df5 = pd.read_csv(io.StringIO(uploaded['example_pandas.csv'].decode('utf-8')))
print("df5 ------------------------------")
print(df5)
```

```
# 01-4-2. 넘파이(Numpy)

import numpy as np
print("numpy : " + np.__version__)

# 1) np.array()
a = np.array([1,2,3,4,5])
print(type(a))
print(a)
print(a.ndim) # 차원 출력
print(a.shape) # 크기 출력
print()

b = np.array([[10, 20, 30], [ 60, 70, 80]]) 
print(b)
print(b.ndim) # 차원 출력
print(b.shape) # 크기 출력
print()

# 2) ndarray의 초기화
a = np.zeros((2,3)) # 모든값이 0인 2x3 배열 생성.
print("np.zeros -------------------------")
print(a)
print()

a = np.ones((2,3)) # 모든값이 1인 2x3 배열 생성.
print("np.ones -------------------------")
print(a)
print()

a = np.full((2,2), 7) # 모든 값이 특정 상수인 배열 생성. 이 경우에는 7.
print("np.full --------------------------")
print(a)
print()

a = np.eye(3) # 대각선으로는 1이고 나머지는 0인 2차원 배열을 생성.
print("np.eye ---------------------------")
print(a)
print()

a = np.random.random((2,2)) # 임의의 값으로 채워진 배열 생성
print("np.random.random ------------------")
print(a)
print()

# 3) np.arange()
a = np.arange(10) #0부터 9까지
print("np.arange ---------------------------")
print(a)
print()

a = np.arange(1, 10, 2) #1부터 9까지 +2씩 적용되는 범위
print("np.arange(start, stop, step) --------")
print(a)
print()

# 4) reshape()
a = np.array(np.arange(30)).reshape((5,6)) # 0부터 n-1 까지의 숫자를 생성하는 arange(n) 함수에 배열을 다차원으로 변형하는 reshape()를 통해 배열을 생성
print("reshape -----------------------------")
print(a)
print()

# 5) Numpy 슬라이싱
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)

b = a[0:2, 0:2]
print(b)

b = a[0, :] # 첫번째 행 출력
print(b)

b=a[:, 1] # 두번째 열 출력
print(b)
print()

# 6) Numpy 정수 인덱싱(integer indexing)
a = np.array([[1,2], [4,5], [7,8]])
print(a)
b = a[[2, 1],[1, 0]] # a[[row2, row1],[col1, col0]]을 의미함.
print(b)
print()

# 7) Numpy 연산
x = np.array([1,2,3])
y = np.array([4,5,6])

print(x)
print(y)
print()

b = x + y # 각 요소에 대해서 더함
# b = np.add(x, y)와 동일함
print("x + y --------------------------")
print(b)
print()

b = x - y # 각 요소에 대해서 빼기
# b = np.subtract(x, y)와 동일함
print("x - y --------------------------")
print(b)
print()

b = b * x # 각 요소에 대해서 곱셈
# b = np.multiply(b, x)와 동일함
print("b * x --------------------------")
print(b)
print()

b = b / x # 각 요소에 대해서 나눗셈
# b = np.divide(b, x)와 동일함
print("b / x --------------------------")
print(b)
print()


a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(a)
print(b)
print()

# 1*5 + 2*7 = 19, 1*6 + 2*8 = 22
# 3*5 + 4*7 = 43, 3*6 + 4*8 = 50

c = np.dot(a, b)
print(c)
```
