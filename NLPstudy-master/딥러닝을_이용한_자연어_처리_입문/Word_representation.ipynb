# 단어의 표현 방법
---

1. 국소 표현(Local Representation)
  - 해당 단어에 특정값을 매핑
  - 예) {"puppy":1, "lovely":2, "cute":3}
2. 연속 표현(Continuous Representation)
  - 해당 단어를 표현하기 위해 주변 단어들을 참고
  - puppy를 cute, lovely한 느낌으로 정의
  - 단어의 뉘앙스 표현이 가능
